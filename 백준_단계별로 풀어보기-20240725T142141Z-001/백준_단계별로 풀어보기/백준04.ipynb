```python
# 04

num_list = []  # 또는 num_list = list()
for i in range(9):
  num_list.append(int(input()))

max_value = num_list[0]
max_index = 0

for i in range(1, len(num_list)):
  if num_list[i] > max_value:
    max_value = num_list[i]
    max_index = i   # max_index = i +1

print(max_value)
print(max_index + 1)    # max_index   => 주석과 같이 코드가 작성되면, index가 0 기반 값이 아닌, 1 기반 값으로 출력되므로 혼동 발생 가능. 따라서 이는 좋지 않은 방식임.
```

```python
# 05

N, M = map(int, input().split())
num_list = []
for i in range(N):
  num_list.append(0)

for ball1 in range(M):
  i, j, k = map(int, input().split())

  for ball2 in range(i - 1, j):
    num_list[ball2] = k

print(" ".join(map(str, num_list)))
```

```python
# 06

N, M = map(int, input().split())
num_list = []
for ball in range(N):
  num_list.append(ball + 1)

for ball_exchange in range(M):
  i, j = map(int, input().split())
  ball_mark = num_list[i - 1]
  num_list[i - 1] = num_list[j - 1]
  num_list[j - 1] = ball_mark

print(" ".join(map(str, num_list)))
```

```python
# 09

N, M = map(int, input().split())
basket_list = list(i + 1 for i in range(N))   # 또는 [i for i in range(N)]

for rev in range(M):
  i, j = map(int, input().split())
  for k in range((j - i) // 2 + 1):
      basket_list[i - 1 + k], basket_list[j - 1 - k] = basket_list[j - 1 - k], basket_list[i - 1 + k]
print(" ".join(map(str, basket_list)))
```

```python
# 10

N = int(input())
score_list = list(map(int, input().split()))
maax = max(score_list)
sum = 0

for i in range(N):
  score_list[i] = score_list[i] / maax * 100
  sum += score_list[i]

print(sum / N)
```
