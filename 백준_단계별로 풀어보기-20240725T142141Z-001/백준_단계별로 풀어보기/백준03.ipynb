```python
# 01

N = int(input())

for i in range(1, 10):
  print(f"{N} * {i} = {N * i}")
```

```python
# 02

T = int(input())

for i in range(0, T):
  A, B = map(int, input().split())
  print(A + B)
```

```python
# 02 - 다르게 출력해보기

T = int(input())
arr = []

for i in range(0, T):
  A, B = map(int, input().split())
  arr.append(A)
  arr.append(B)   # arr.append(A, B) 이런 형태의 함수 사용은 불가능.

for i in range(0, 2 * T, 2):    # C와 마찬가지로 2T 불가능. 2 * T 가능.
  print(arr[i] + arr[i + 1])
```

```python
# 03

n = int(input())
sum = 0

for i in range(0, n):
  sum += (i + 1)

print(sum)
```

```python
# 04

X = int(input())
N = int(input())
sum = 0

for i in range(0, N):
  a, b = map(int, input().split())
  sum += (a * b)

if X == sum:
  print("Yes")
else:
  print("No")
```

```python
# 05

N = int(input())

for i in range(0, (N // 4)):
  print("long", end=" ")

print("int")
```

```python
# 06 - 실행은 안되는데 답 맞음

import sys

T = int(sys.stdin.readline())

for i in range(T):
  A, B = map(int, sys.stdin.readline().split())
  print(A + B)
```

```python
# 09

N = int(input())

for i in range(0, N):
  print('*' * (i + 1))
```

```python
# 10

N = int(input())

for i in range(0, N):
  print(" " * (N - (i + 1)), end= "")
  print("*" * (i + 1))
```

```python
# 10 - 같은 출력 다른 방식

N = int(input())

for i in range(0, N):
  print(" " * (N - (i + 1)) + '*' * (i + 1))
```

```python
# 11

while True:
  sum = 0

  A, B = map(int, input().split())
  sum = A + B

  if A == 0 and B == 0:
    break

  print(sum)
```

```python
# 12 - 정수 이외에 다른 입력이 들어오면 eror가 발생하며 반복문이 종료되게 함

while True:
  try:
    A, B = map(int, input().split())
    print(A + B)
  except:
    break
```

```python
# 12 - 이거도 가능하다

import sys

for i in sys.stdin:
  A, B = map(int, i.split())
  print(A + B)
```
