```python
# 06

S = input()

alp = list(-1 for i in range(26))   # 또는 alp = [-1] * 26

for i in range(len(S)):
  if alp[ord(S[i]) - 97] == -1:
    alp[ord(S[i]) - 97] = i

print(" ".join(map(str, alp)))
```

```python
# 06 - 다른 방법으로 다시 풀어보기

alp = "abcdefghijklmnopqrstuvwxyz"
text = input()

for i in alp:
  if i in text:
    print(text.index(i), end=" ")   # 문자가 처음 등장하는 위치만을 반환한다. / text의 index를 반환함.
  else:
    print(-1, end=" ")
```

```python
# 07

T = int(input())

for i in range(T):
  R, S = input().split()
  for rep in S:   # 이런 형태의 문자열 접근 방식은 rep의 전,후 요소에 접근할 수는 없다.
    print(int(R) * rep, end="")
  print()
```

```python
# 08

text = input()
count = 0

if text[0] != " ":
  count += 1

for i in range(len(text) - 1):
  if text[i] == " ":
    if text[i + 1] != " ":
      count += 1

print(count)
```

```python
# 09

A, B = input().split()

def rev(num):
  num_list = list(num)    # 문자열은 불변형이므로 리스트로 변환하여 값을 뒤집어준다.

  for i in range(len(num_list) // 2):
    num_list[i], num_list[-(i + 1)] = num_list[-(i + 1)], num_list[i]

  return "".join(num_list)    # 리스트를 문자형으로 변환

# rev 함수의 반환값을 변수에 저장
rev_A = rev(A)
rev_B = rev(B)

# rev(A) / 이는 함수의 반환값을 출력하는 것이 아닌, 단순히 함수만 호출하는 것이다.
# 따라서 print(rev(A)) 해주어야함.
if rev_A > rev_B:
    print(rev_A)
else:
    print(rev_B)
```

```python
# 10

alp_list = ["ABC", "DEF", "GHI", "JKL", "MNO", "PQRS", "TUV", "WXYZ"]
phone_alp = input()
sum = 0
for i in phone_alp:
  for group in alp_list:
    if i in group:
      sum += alp_list.index(group) + 3

print(sum)
```

```python
# 09 - 다시 풀어보기

A, B = input().split()

def rev(num):
  num_list = list(num)
  for i in range(len(num) // 2):
    num_list[i], num_list[-(i + 1)] = num_list[-(i + 1)], num_list[i]
  return "".join(num_list)

print(rev(A))
print(rev(B))
```

```python
# 10 - 다시 풀어보기

num_list = ["ABC", "DEF", "GHI", "JKL", "MNO", "PQRS", "TUV", "WXYZ"]
mine_list = list(input())
sum = 0

for text in mine_list:
  for alp in num_list:
    if text in alp:
      sum += num_list.index(alp) + 3

print(sum)
```
