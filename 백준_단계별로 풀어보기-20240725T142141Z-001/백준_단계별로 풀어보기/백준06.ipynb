```python
# 02 - 리스트를 사용하여 풀어보기

num_list = [1, 1, 2, 2, 2, 8]
mine_list = list(input().split())     # 이는 mine_list = input().split()와 동일하다. / 입력받은 문자열을 리스트로 반환하여 변수에 저장.

for i in range(len(num_list)):
  num_list[i] -= int(mine_list[i])

print(" ".join(map(str, num_list)))   # print(" ".join(num_list)) 를 쓰는 경우는 join 안의 리스트가 원래 문자형이었다면 가능.
                                      # 이 문제는 문자형이 아닌, 정수형이었기 때문에 위와 같이 출력해주어야 함.
```

```python
# 03

N = int(input())

for i in range(N):
  print(" " * (N - (i + 1)), "*" * (2 * i + 1))
for j in range(N - 1, 0, -1):
  print(" " * (N - j), "*" * (2 * j - 1))
```

```python
# 04

word = list(input())
mark = "".join(word)

for i in range(len(word) // 2):
  word[i], word[-(i + 1)] = word[-(i + 1)], word[i]

word_str = "".join(word)

if mark == word_str:
  print(1)
else:
  print(0)
```

```python
# 05

text = input()
new_text = ""
alp = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
alp_list = [0] * 26

for i in text:    # 입력받은 문자열을 모두 대문자로 출력해주기 위한 반복문
  if 97 <= ord(i) <= 122:
    new_text += chr(ord(i) - 32)
  else:
    new_text += i

for i in new_text:    # 알파벳 횟수 카운팅
  alp_list[alp.index(i)] += 1

if alp_list.count(max(alp_list)) >= 2:    # 가장 많이 나온 알파벳이 2가지 이상일 때
  print("?")
else:
  print(alp[alp_list.index(max(alp_list))])
```

```python
# 05 - 다른 방법(첫 방법은 시간 초과뜸.)

text = input().upper()    # 입력받은 문자를 모두 대문자로 바꿈 / 소문자로 바꾸는 메서드는 lower()
text_list = list(set(text))   # 문자의 중복제거. 이를 하는 이유는 맨 마지막줄 코드
lst = []

for i in text_list:
  cnt = text.count(i)
  lst.append(cnt)

if lst.count(max(lst)) >= 2:
  print("?")
else:
  print(text_list[lst.index(max(lst))])   # index를 통해 해당 문자를 출력하기 위함.
```

```python
# 06

cro_list = ["c=", "c-", "dz=", "d-", "lj", "nj", "s=", "z="]
cro_alp = input()

for i in cro_list:
  cro_alp = cro_alp.replace(i, "*")   # replace 메서드는 문자형에서 사용됨.

print(len(cro_alp))
```

```python
# 07

N = int(input())
text_count = N

for i in range(N):
  text = input()
  for j in range(len(text) - 1):
    if text[j] == text[j + 1]:
      continue
    elif text[j] in text[j + 2:]:   # 슬라이싱 / slice => sequence[start:stop:step]
      text_count -= 1
      break

print(text_count)
```

```python
# 08

grade_dic = {"A+" : 4.5, "A0" : 4.0, "B+" : 3.5, "B0" : 3.0, "C+" : 2.5, "C0" : 2.0, "D+" : 1.5, "D0" : 1.0, "F" : 0.0}
sum_mul = 0
sum_grade = 0

for i in range(20):
  sub, grade, score = input().split()
  if score != "P":
    mul = float(grade) * grade_dic.get(score)
    sum_mul += mul
    sum_grade += float(grade)

print(sum_mul / sum_grade)
```
