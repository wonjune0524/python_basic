```python
# 01

A, B = input().split()

if int(A) > int(B):
  print(">")
elif int(A) == int(B):
  print("==")
else:
  print("<")
```

```python
# 02

score = int(input())

if 90 <= score <= 100:
  print('A')
elif 80 <= score <= 89:
  print('B')
elif 70 <= score <= 79:
  print('C')
elif 60 <= score <= 69:
  print('D')
else:
  print('F')
```

```python
# 03

year = int(input())

if year % 4 ==0:
  if year % 400 == 0:
    print(1)
  elif year % 100 != 0:
    print(1)
  else:
    print(0)
else:
  print(0)
```

```python
# 04

x = int(input())
y = int(input())

if x > 0 and y > 0:
  print(1)
elif x < 0 and y > 0:
  print(2)
elif x < 0 and y < 0:
  print(3)
else:
  print(4)
```

```python
# 05

H, M = input().split()

h = int(H)
m = int(M)

if m < 45:
  if h == 0:
    h = 23
  else:
    h -= 1
  m += 15
  print(f"{h} {m}")

else:
  m -= 45
  print(f"{h} {m}")
```

```python
# 06

A, B = input().split()
C = input()

a = int(A)
b = int(B)
c = int(C)
M = b + c
H = a + (M // 60)

if M >= 60:
  if H >= 24:
    H = H % 24

  M = M % 60
  print(f"{H} {M}")

else:
  print(f"{H} {M}")
```

```python
# 07

num1, num2, num3 = map(int, input().split())

if num1 == num2 == num3:
  print(10000 + num1 * 1000)
elif num1 == num2:
  print(1000 + num1 * 100)
elif num2 == num3:
  print(1000 + num2 * 100)
elif num3 == num1:
  print(1000 + num3 * 100)
else:
  if num1 > num2 and num1 > num3:
    print(num1 * 100)
  elif num2 > num1 and num2 > num3:
    print(num2 * 100)
  else:
    print(num3 * 100)
```
