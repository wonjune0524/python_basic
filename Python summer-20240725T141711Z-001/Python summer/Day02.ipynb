```python
# if문 예제
# 입력받은 숫자가 5 이상인지, 10 이상인지 판별하기

# 여러줄 주석 : '''(작은 따옴표 3번)
'''
if 조건문 :     # 괄호도 사용할 수 있음. ex) if () :
  #
elif 조건문 :   # else if 와 동일.
  #
else :
  #
'''
number1 = int(input("숫자를 입력해 주세요!\n"))

if (number1 >= 10):
  print("이 숫자는 10보다 크거나 같습니다.")
elif (number1 >= 5):
  print("이 숫자는 5보다 크거나 같습니다.")
else :
  print("이 숫자는 5보다 작습니다.")
```

```python
# 1부터 50까지 숫자 중에서
# 3의 배수이거나 5의 배수인 숫자의 개수를 출력해보기

count = 0
for i in range(1, 51):
  if (i % 3 == 0 or i % 5 == 0):
    print(i)
    count += 1    # count++는 파이썬에서 지원 안됨.
print(f"해당하는 숫자의 개수는 {count}가지이다.")
```

```python
# 소수인지 판별
num = 29

isPrime = True
if (num > 1):
  for i in range(2, int(num ** 0.5) + 1):
    if (num % i == 0):
      isPrime = False
      break

if isPrime:
  print("소수입니다!")
else:
  print("소수가 아닙니다!")
```

```python
# 소수인지 판별

while True:
  isPrime = True

  num = int(input("정수를 입력해주세요! (프로그램을 종료하고 싶으시면 0을 입력하세요.): "))
  if (num == 0):
    print("프로그램이 종료되었습니다.")
    break

  elif (num > 1):
    for i in range(2, int(num ** 0.5) + 1):
      if (num % i == 0):
        isPrime = False
        break

  if isPrime:
    print("소수입니다!")
  else:
    print("소수가 아닙니다!")
```

```python
# 비트연산자
# 정수형 숫자에 대해서 이진수 단위로 바로 접근해서 연산을 수행하는 연산자

# AND(&)
# OR(|)
# XOR(^)
# NOT(~)
# 시프트 연산(<<, >>)
```

```python
# AND 연산자 (&)
a = 60    # 2진수 표현법 => 111100  / 60을 2로 쭉 나누었을 때의 나머지 거꾸로 나열
b = 13    # 2진수 표현법 => (00)1101    / (2^3 * 1) + (2^2 * 1) + (2^1 * 0) + (2^0 * 1) = 13
result = a & b
# 비트연산에서 이진수 한 자리가 1이면 True, 0이면 False
# & (and) 연산자는 둘 다 1이어야 True
'''
111100
001101
------
001100    => 둘 다 1이어야 True. 따라서 이는 (2^3 * 1) + (2^2 * 1) = 12
'''

print(result)
```

```python
# OR 연산자(|)
result = a|b
print(result)
```

```python
# XOR 연산자(^)
result = a^b
print(result)
# exclusive-or은 두 비트가 서로 같을 때 0, 서로 다를 때 1을 반환
```

```python
# NOT 연산자
prac = True
if not prac:      # Not True => False
  print("Here")
else:
  print("There")
```

```python
# 비트 NOT 연산자(~)
# 비트를 0에서 1로, 1에서 0으로 반전시킴
b = 13
result = ~b
print(result)
```

```python
# 시프트 연산
# 비트를 왼쪽이나 오른쪽으로 이동시킴
# 빈 자리는 0으로 메꿈
a = 60
result1 = a<<2
print(result1)
print(result>>2)
result2 = a>>2
print(result2)
```

```python
num = 0b1010    # 0b 뜻 => 숫자를 이진수 그 자체로 출력
bit_to_set = 0b0100

result = num | bit_to_set
print(bin(result))
```

```python
num = 0b1010
bit_to_clear = 0b0010

result = num & ~bit_to_clear
print(result)
print(bin(result))
```
