```python
# example.py

import requests   # command prompt => pip install requests
from bs4 import BeautifulSoup as bs   # pip install BeautifulSoup4
# requests 모듈 : url로 웹페이지의 내용을 가져옴
# BeautifulSoup : 웹페이지 내용을 좀 더 정갈하게 다듬어주는 모듈

# 예시 url : 'https://sports.news.naver.com/kbaseball/record/index?category=kbo'
url = 'https://sports.news.naver.com/kbaseball/record/index?category=kbo'

response = requests.get(url)
print(response.status_code)   # 상태 코드 출력 / 이는 200 이 출력될 때만 성공이다.

html = response.text
print(html)
```

```python
soup = bs(html, 'html.parser')    # BeautifulSoup 모듈로 html 파싱
print(soup)
```

```python
ranking_table = soup.find('tbody', id = 'regularTeamRecordList_table')
# tbody tag이면서, 해당 id 속성을 가지는 요소들을 불러옴
print(ranking_table)
```

```python
row = ranking_table.find_all('tr')
for r in row:
  print(row)
```

```python
for r in row:
  th = r.find('th').text.strip()
  div = r.find_all('div')[0].text.strip()
  print(f"{th}위 {div}")

for r in row:
  div = r.find_all('div')[0].text.strip()
  td = r.find_all('td')[5].text
  print(f"{div} 승률 {td}")
```

```python
# https://www.melon.com/chart/index.htm
import requests
from bs4 import BeautifulSoup as bs

url = "https://www.melon.com/chart/index.htm"

# 헤더 설정
headers = {
    'Accept' : 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers)
print(response.status_code)
```

```python
soup = bs(response.text, 'html.parser')
# response.text 웹사이트 내의 html 태그들
# BeautifulSoup parsing

songs = soup.find_all('tr', class_ = 'lst50')    # lst50 이름으로 class 속성을 가지는 tr 태그들의 집합
print(songs)
```

```python
song_list = []
for tr in songs:
  rank_number = tr.find_all('span', class_= 'rank')[0].text
  rank_text = tr.find_all('div', class_ = 'wrap_song_info')[0]
  title = rank_text.find_all('div', class_ = 'rank01')[0].text.strip()
  singer = rank_text.find_all('div', class_ = 'rank02')[0].find('a').text.strip()
  print(f"{rank_number} : {title} - {singer}")
  song_list.append([rank_number, title, singer])    # list 안에 list로 형
```

```python
import csv
f = open('/content/drive/MyDrive/Python summer/melon.csv', 'w', newline = "", encoding = "utf-8")
# 이 경로의 파일에 접근해서, utf-8 인코딩 방식으로 파일 내에 요소를 쓰겠다
writer = csv.writer(f)
col_name = ['순위', '노래 제목', '가수명']    # table column명을 설정
writer.writerow(col_name)   # 행 하나 추가
writer.writerows(song_list)
f.close()
```
