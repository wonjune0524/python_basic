연산자

산술연산자
- 덧셈 +
- 뺄셈 -
- 곱셈 *
- 나눗셈 /
- 몫 //
- 나머지 %
- 거듭제곱 ** # 파이썬에서 특별한 연산자
- 대입 =

```python
num1 = input('num1을 입력하세요\n')
num2 = int(input('num2를 입력하세요\n'))
num1 = int(num1)
```

```python
print(num1 + num2)
print(num1 - num2)
print(num1 * num2)
print(num1 / num2)    # 실수 리턴
print(num1 // num2)   # 정수 리턴 : 몫
print(num1 % num2)    # 정수 리턴 : 나머지
print(num1 ** num2)
```

```python
# 단순대입
num1 = 5
num2 = 10
num1 = 5 + num2
print(num1)
```

```python
# 복합대입 연산자
num1 = num1 + 5
print(num1)

for i in range(1, 5):
  num1 += 5
  print(num1)
```

```python
num1 -= 7
print(num1)
num1 *= 11
print(num1)
```

```python
sum = 1 + 2 + 3 + 4 + 5
print("결과는", sum, "입니다.")
```

```python
# f-string
hour = 1.4
print(f'나는 집까지 가는 데 {hour}시간이 걸립니다')
print(f'나는 집까지 가는 데 {hour:.2f}시간이 걸립니다')
```

```python
num1 = 10
num2 = 3
print(num1 // num2)
print(num1 % num2)
# 아래에
# 10을 3으로 나눈 몫은 3, 나머지는 1입니다 출력
print(f'{num1}을 {num2}로 나눈 몫은 {num1//num2}, 나머지는 {num1%num2}입니다.')
```

논리 연산자
- and, or, not

```python
print(10 > 0 and 5 > 0)
```

```python
num1 = 10
num2 = 20

print(num1 > num2)
print(num1 >= num2)
print(num1 < num2)
print(num1 <= num2)
print(num1 != num2)   # 같지 않다
print(num1 == num2)   # 같다
```

```python
# 논리연산자
number1 = 10
number2 = 20
print(number1 > 0 and number2 > 0)    # and 연산자는 모든 조건이 True여야 True
print(number1 < 0 and number2 < 0)
print(number1 > 0 or number2 > 0)     # or 연산자는 모든 조건이 False여야 False
print(number1 < 0 or number2 < 0)
print(number1 > 0 and number2 < 0)
print(number1 > 0 or number2 < 0)
```
