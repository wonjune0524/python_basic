```python
# 6 : 단어 빈도 수 계산하기
input_str = "this is a test this is only a test"

words = input_str.split()
# split 함수는 구분자를 기준으로 문자열을 잘라서 리스트로 return
print(words)
```

```python
for word in words:
  print(word)
```

```python
word_freq = {}
for word in words:
  if word in word_freq:
    word_freq[word] += 1
  else:
    word_freq[word] = 1

print(word_freq)
```

```python
# 7 : 각 학생들의 성적이 주어졌을 때
# 각 학생들의 평균점수를 계산
grades = {
    "Alice": [85, 90, 92],
    "Bob" : [78, 81, 85],
    "Charlie" : [92, 95, 77]
}

averages = {}

for k in grades:
  print(f"{k}의 점수들은 {grades.get(k)}입니다.")   # 또는 grades[k]

for k, v in grades.items():   # 함수 items()는 key와 value를 동시에 return.
  print(f"{k}의 점수들은 {v}입니다.")
```

```python
for student, scores in grades.items():
  averages[student] = sum(scores) / len(scores)

print(averages)
```
