```python
# 유효한 괄호 문자열인지 확인하는 프로그램

# s = "(){}[]" : True
# s = "([)]" : False
# s = "{[]}" : True
s = "()[]{}"

lst = []    # 이는 False를 의미함.
mapping = { ")":"(", "}": "{", "]": "["}    # key : value 형태임.

for char in s:
  if char in mapping:   # key 일 때
    if lst:   # lst에 뭐가 담겨 있으면 True로 이 조건문이 출력됨.
      top = lst.pop()
    else:
      top = "#"
    if mapping[char] != top:
      print(False)  # 마지막까지 여기서 안걸리면 True 출력
      break
  else:                 # value 일 때
      lst.append(char)
else:
  print(not lst)
```

```python
# 가변 매개변수 : 변수가 몇 개 들어올지 모를 때
def showNumber(*args):    # tuple 형태로 반환
  print("args", args)
  print("args type", type(args))
  for i in args:
    print(i)

showNumber(1, 2, 3, 4, 5)
```

```python
def greet(greeting, *args):   # 가변 parameter는 일반 변수와 같이 사용시 맨 뒤에 위치해야 한다.
  for name in args:
    print(f"{greeting}, {name}!")

greet("Hello", "길동", "둘리", "또치")
```

```python
# 가변 parameter 활용
def get_mx(*numbers):
  max = -1
  for i in numbers:
    if i > max:
      max = i
  return max

def get_mn(*numbers):
  min = 999999
  for i in numbers:
    if i < min:
      min = i
  return min

print(get_mx(1, 11, 41, 55, 34))
print(get_mn(1, 11, 41, 55, 34))
```

```python
# 입력받은 숫자들의 합과 평균을 동시에 return 하는 함수
# len(tuple) <--

def sum_avg(*numbers):
  sum = 0
  for i in numbers:
    sum += i
  return sum, sum / len(numbers)
"""
x, y = sum_avg(1, 2, 3, 4, 5, 6, 7)
print(x, y)
"""
print(sum_avg(1, 2, 3, 4, 5, 6, 7))
```

```python
class Person:   # class 정의
  def __init__(self, name, age):    # 생성자라고 하는 class 내부 함수, 함수 내부의 첫번째 변수는 꼭 self로 지정해야함.
    self.name = name
    self.age = age
  def greet(self):
    print(f"Hello, my name is {self.name} and my age is {self.age}")

p1 = Person("Alice", 30)
p2 = Person("Bob", 25)
print(f"{p1.name}의 나이는 {p1.age}이다")
print(f"{p2.name}의 나이는 {p2.age}이다")
p1.greet()
p2.greet()
```

```python
class Person:   # class 정의
  species = "Homo sapiens"    # class 변수 : 모든 객체가 동일하게 공유
  def __init__(self, name, age):    # 생성자라고 하는 class 내부 함수
    self.name = name    # instance 변수
    self.age = age      # 객체를 생성할 때마다 달라지는 변수
  def greet(self):
    print(f"Hello, my name is {self.name} and my age is {self.age}")

p1 = Person("Alice", 30)
p2 = Person("Bob", 25)
print(p1.species)
print(p2.species)
```

```python
class WareHouse:
  stock_num = 0
  def __init__(self, name):
    self.name = name
    WareHouse.stock_num += 1  # class 변수에 접근
  def __del__(self):  # 소멸자
    WareHouse.stock_num -= 1

w1 = WareHouse("Incheon")
w2 = WareHouse("Seoul")
print(WareHouse.stock_num)
print(w1.stock_num)
print(w1.name)
del w1
print(WareHouse.stock_num)
```

```python
class Korean:
  country = "한국"
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def __del__(self):
    print("destroy the object")
  def yell(self):
    print(f"{self.name}이(가) 외쳤다!")

k1 = Korean("민수", 50)
print(k1.name)
print(k1.age)
print(k1.country)
print(Korean.country)
k1.yell()
del k1
```

```python
# 상속
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def greet(self):
    print(f"Hello, my name is {self.name}")

class Student(Person):
  def __init__(self, name, age, subject):
    super().__init__(name, age)
    self.subject = subject
  def study(self):
    print(f"{self.name} 학생은 {self.subject} 공부를 합니다")

s = Student("Alice", 20, "math")
s.greet()
s.study()
p = Person("John", 25)
p.greet()
```

```python
# 오버라이딩
class Animal:
  def sound(self):
    raise NotImplementedError("상속받은 클래스는 이걸 구현하세요!")

class Dog(Animal):
  def sound(self):
    print("멍멍")

class Cat(Animal):
  def sound(self):
    print("야옹")

dog = Dog()
cat = Cat()
dog.sound()
cat.sound()
```
